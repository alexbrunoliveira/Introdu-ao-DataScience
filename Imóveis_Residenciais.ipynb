```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/drive/MyDrive/Curso Pandas/aluguel.csv', sep = ';')
```

```python
dados.head(10)
```

```python
# Eliminando dados 'Tipo' duplicados
list(dados['Tipo'].drop_duplicates())
```

```python
# criando lista
residencial = ['Quitinete',
'Casa',
'Conjunto Comercial/Sala',
'Apartamento',
'Casa de Condomínio',
'Prédio Inteiro',
'Flat',
'Loja/Salão',
'Galpão/Depósito/Armazém',
'Casa Comercial',
'Casa de Vila',
'Terreno Padrão',
'Box/Garagem',
'Loft'
'Loja Shopping/ Ct Comercial',
'Chácara',
'Loteamento/Condomínio',
'Sítio',
'Pousada/Chalé',
'Studio',
'Hotel',
'Indústria']
```

```python
# Selecionando apenas dados residenciais
residencial = ['Quitinete', 
'Casa',
'Apartamento',
'Casa de Condomínio',
'Casa de Vila']
```

```python
# metodo isin, usado para verificar se os elementos existem com true e false dentro do dataFrame
dados['Tipo'].isin(residencial).head(10)
```

```python
# Atribuindo a lista de imoveis a seleção
selecao = dados['Tipo'].isin(residencial)
```

```python
# Metodo de seleção no data Frame e trazendo dados apenas da inf. true
dados_residencial = dados[selecao]
```

```python
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
#checando o tamanho dataFrame
dados_residencial.shape[0]
```

```python
# construcao do indice
dados_residencial.index = range(dados_residencial.shape[0])
```

```python
dados_residencial
```
